```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
import json # library to handle JSON files
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

venues = results['response']['venues']  
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['name', 'categories', 'location.lat', 'location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head(10)
```

# Let's create a function to repeat the same process for all the institites in Toronto

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url='https://api.foursquare.com/v2/venues/search?categoryId={}&limit={}&radius={}&client_id={}&client_secret={}&ll={},{}&v={}'.format(
            categoryId,limit,radius,CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION)
           
        # make the GET request
        results = requests.get(url).json()['response']['venues']
       # ['response']['venues']
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['name'], 
            v['location']['lat'], 
            v['location']['lng'],  
            v['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['institite', 
                  'institite Latitude', 
                  'institite Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### Now write the code to run the above function on each institite and create a new dataframe called *institute_venues*.

## 4. <a id="item4"></a>  Analyze venues in each institute 

```python
# one hot encoding
toronto_onehot = pd.get_dummies(institute_venue[['Venue_Category']], prefix="", prefix_sep="")

# add institute column back to dataframe
toronto_onehot['institute'] = institute_venue['institute'] 

# move institute column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot
```

And let's examine the new dataframe size.

```python
toronto_onehot.shape
```

#### Next, let's group rows by institute and by taking the sum of the occurrence of each venue category

```python
toronto_onehot_sum = toronto_onehot.groupby('institute').sum().reset_index()
```

```python
toronto_onehot_sum.head()
```

#### Next, let's group rows by institute and by taking the mean of the frequency of occurrence of each venue category

```python
toronto_grouped = toronto_onehot.groupby('institute').mean().reset_index()
```

```python
toronto_grouped.to_csv('toronto_grouped.csv',encoding = 'utf-8')
```

#### Let's confirm the new size

```python
toronto_grouped.head()
```

#### Let's print each institute along with the top 5 most common venues

```python
num_top_venues = 5

for hood in toronto_grouped['institute']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['institute'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### Let's put that into a *pandas* dataframe

First, let's write a function to sort the venues in descending order.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

### Now let's create the new dataframe and display the top 10 venues for each Institute.

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['institute']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
institutes_venues_sorted = pd.DataFrame(columns=columns)
institutes_venues_sorted['institute'] = toronto_grouped['institute']

for ind in np.arange(toronto_grouped.shape[0]):
    institutes_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

institutes_venues_sorted
```

### Save data for later processing

```python
institutes_venues_sorted.to_csv('institute_veneues_rank.csv',encoding = 'utf-8')
```

## 5. <a id="item5"></a> Cluster Institutes

#### Run *k*-means to cluster the institutes into 5 clusters.

```python
toronto_onehot_sum=toronto_onehot_sum.set_index('institute')
toronto_onehot_sum
```

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans
# run k-means clustering
kmeans = KMeans(n_clusters = 5, random_state = 0).fit(toronto_onehot_sum)
```

```python
kmeans.labels_
```

### Find best clusters 

```python
means_df = pd.DataFrame(kmeans.cluster_centers_)
means_df.columns = toronto_onehot_sum.columns
means_df.index = ['C1','C2','C3','C4','C5']
means_df['Total Sum'] = means_df.sum(axis = 1)
means_df.sort_values(axis = 0, by = ['Total Sum'], ascending=False)
```

### Label institutes with clusters

```python
cluster_summary = pd.DataFrame([toronto_onehot_sum.index, 1 + kmeans.labels_]).T
cluster_summary.columns = ['institute', 'Cluster']
cluster_summary
```
